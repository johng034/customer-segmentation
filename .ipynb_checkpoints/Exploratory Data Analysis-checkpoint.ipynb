# Exploratory Data Analysis
In this notebook, we will be exploring the dataset. 

The following is a list of what we will be looking for.
- Find any missing values and decide what to do with rows with missing values, if any
- Identify any outliers in the dataset
- View the unique values in each column
- Think about creating new columns that may be more helpful for modeling

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import missingno.missingno as msno
```

```python
# Read the dataset
df = pd.read_csv('marketing_campaign.csv')

df.head()
```

```python
df.describe()
```

## Missing Values

```python
msno.matrix(df)
plt.show()
```

```python
missing_rows = df.Income.isna().sum()
total_rows = df.shape[0]

print(f"""
There are {missing_rows} missing data points in the "Income" column.
That makes up {round((missing_rows/total_rows)*100, 2)}% of the total dataset.
""")
```

```python
df[df.isnull().any(axis=1)].head(10)
```

### Action
Since the number of rows with missing values makes up a small percentage of the total rows, we will simply remove those rows for our analysis.

```python
df.dropna(axis=0, inplace=True)
```
